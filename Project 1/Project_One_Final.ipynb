    Write a function that takes in values and returns the mean of the values. Create a list of numbers that you test on your function to check to make sure your function works!

```python
def getmean(sample):
    number = 0 
    for num in sample:
        number = num + number
        
    return (number/len(sample))

getmean([1,2,4,5,7,2,3,43])
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
def getstdev(sample):
    number = 0
    for num in sample:
        number = number + ((num-getmean(sample))**2)
    return ((number)/len(sample))**0.5
```

```python
def check_dups_missing(df, col, ref_state, ref_col):
    get_not_in_us = df[df[col].apply(lambda x: not check_if_USstate(x, ref_state, ref_col))]
    get_duplicates = df[df.duplicated()]
    
    missingdata = len(get_not_in_us)
    dup = len(get_duplicates)
    
    print(f'Df has {missingdata} of missing data')
    print(f'Df has {dup} of duplicated data\n')
    
    if missingdata != 0:
        print(f'\
Array of missing data under column: "{col}"\n\
Missing data array: {get_not_in_us[col].values} \
at Index: {get_not_in_us[col].index.values}\n\
        ')
    else:
        print('No missing dataframe\n')
    
    if dup != 0:
        print(f'\
Array of duplicated data under column: "{col}" \n\
Duplicated words: {get_duplicates[col].values} \
at Index: {get_duplicates.index.values}\
        ')
    else:
        print('No duplicated dataframe')
```

```python
# define function that tells us difference between two strings under one column
## we reference index as it will not have issues calling
### function considers (number of chars) and (the char by incrementing index)

def get_char_diff(df, col, L_index, R_index):
    # set variable to store length of char for left and right 
    # need to call out values else will return series 
    # need to call out [0] for values else will return numpy array
    L_numchar = len(df.loc[[L_index],col].values[0])
    R_numchar = len(df.loc[[R_index],col].values[0])
    
    # if both numbers the same -->> for loop 
    if L_numchar == R_numchar:
        
        # enumerate to call same index for right table
        for num, left_letter in enumerate (df.loc[[L_index],col].values[0]):
            right_letter = df.loc[[R_index],col].values[0][num]
            if right_letter != left_letter:
                print(f'Left list has char: "{left_letter}" at column index: {num}')
                print(f'Right list has char: "{right_letter}" at column index: {num}')
    else:
        print(f'Left data has {L_numchar} characters')
        print(f'Right data has {R_numchar} characters')
```

```python
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels, sharex=False, kde=False, figsize=(14,7)):
    nrows = int(np.ceil(len(list_of_columns)/2)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=figsize, sharex=sharex, sharey=True)
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    
    for i, column in enumerate(list_of_columns):
        # Calculate mean
        mean = np.mean(dataframe[list_of_columns[i]])
        ax[i].axvline(mean, ls='--', color='crimson')
        
        # Set color of graph according to test type
        if 'act_' in column:
            color = '#E1812C'
        elif 'sat_' in column:
            color = '#3274A1'
        
        sns.histplot(data=dataframe, x=column, color=color, ax=ax[i], kde=kde, bins=9)
        ax[i].set_title(list_of_titles[i] + f' (Mean: {mean:.1f})')
        ax[i].set_xlabel(list_of_xlabels[i])
        # To prevent label duplication
        if i % 2 == 0:
            ax[i].set_ylabel('Number of States')
    
    fig.tight_layout(h_pad = 1.5, w_pad = 1.5)
```

```python
def plot_hist_against_ave(df,col,title_name,x_axis,y_axis):
    a = final_income[col].mean()
    df = final_income[['state',col]]

    df['comp_diff'] = df[col].map(lambda x: ((x-a)/a) * 100)

    df = df.sort_values('comp_diff')

    x = df['comp_diff']
    y = df['state']

    # establish figure size 
    ax = plt.figure(figsize = (17,17))
    
    # label axis 
    plt.title(title_name, fontsize = 20)
    plt.xlabel(f'{x_axis} (Mean of {round(a,2)})', fontsize = 20)
    plt.ylabel(y_axis, fontsize = 25)

    # plot barh 
    plt.barh(y, x)    
```
